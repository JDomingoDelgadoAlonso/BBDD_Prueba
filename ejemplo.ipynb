```python
# Paso 1: Importar bibliotecas
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split

# Paso 2: Crear los datos (tama√±o de la casa en metros cuadrados y precios)
# Datos simulados
metros_cuadrados = np.array([10, 30, 50, 60, 90, 100, 110, 150, 170, 200, 240]).reshape(-1, 1)
precios = np.array([140000, 142000, 150000, 160500, 160000, 170000, 172000, 180000, 190090, 200000, 400000])

# Paso 3: Dividir los datos en conjuntos de entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(metros_cuadrados, precios, test_size=0.2, random_state=42)

# Paso 4: Crear el modelo de regresi√≥n lineal y entrenarlo
modelo = LinearRegression()
modelo.fit(X_train, y_train)

# Paso 5: Realizar predicciones
predicciones = modelo.predict(X_test)

# Mostrar los coeficientes y el intercepto
print(f"Coeficiente de regresi√≥n (pendiente): {modelo.coef_[0]}")
print(f"Intercepto: {modelo.intercept_}")

# Paso 6: Gr√°fica con los datos y la l√≠nea de regresi√≥n
plt.figure(figsize=(8, 6))
plt.scatter(metros_cuadrados, precios, color='blue', label='Datos reales')
plt.plot(metros_cuadrados, modelo.predict(metros_cuadrados), color='red', label='L√≠nea de regresi√≥n')

# Etiquetas y leyenda
plt.title('Regresi√≥n Lineal: Predicci√≥n del Precio de una Casa')
plt.xlabel('Tama√±o de la casa (m¬≤)')
plt.ylabel('Precio (USD)')
plt.legend()
plt.grid(True)
plt.show()

# Paso 7: Predicci√≥n para una casa de 125 metros cuadrados
metros_nueva_casa = np.array([[125]])
precio_predicho = modelo.predict(metros_nueva_casa)
print(f"El precio predicho para una casa de 125 metros cuadrados es: ${precio_predicho[0]:,.2f} USD")
```
